# **Análise de desempenho nas áreas do Enem dos alunos aprovados pelo SISU no curso de Ciência da Computação na UFCG**

O Sistema de Seleção Unificada (SISU) é um processo seletivo eletrônico criado e gerido pelo Ministério da Educação, no qual instituições públicas de ensino superior de todo o Brasil ofertam vagas para os candidatos que realizaram o Exame Nacional do Ensino Médio (Enem), avaliação composta por 180 questões distribuídas em 4 áreas de conhecimentos gerais e um texto dissertativo-argumentativo. O processo ocorre duas vezes ao ano, geralmente nos meses de março e julho, e ao final do período das inscrições, os estudantes são selecionados por curso e modalidade de concorrência baseado na classificação e na disponibilidade das vagas ofertadas. A inscrição é gratuita, em uma única etapa e feita pela Internet, pelo Portal Único de Acesso ao Ensino Superior.

A Universidade Federal de Campina Grande (UFCG) é reconhecida pelo alto desempenho e avaliação dos padrões MEC para o ensino superior, principalmente no curso de Ciência da Computação, que conta com uma [avaliação de 5 estrelas](https://www.computacao.ufcg.edu.br/graduacao/procedimentos-graduação/avaliação). Em consonância ao SISU, são ofertadas 180 vagas por ano (distribuídas igualmente nas duas edições) para o curso de Computação na instituição. A partir desse contexto, surgem questionamentos acerca da distribuição do desempenho nas áreas do Enem dos candidatos classificados pelo SISU. Com o objetivo de sanar tais questões, foi desenvolvida a presente análise com o intuito de compreender as variações nas notas entre os períodos de 2018.1 à 2022.1.

Com o intuito de especificar a análise, as databases foram filtradas, unicamente, para os candidatos que concorreram pela Ampla Concorrência, a fim de evitar grandes oscilações dos dados das notas entre as diversas modalidades de concorrência, tendo em vista que cada uma possui uma nota de corte distinta.

O primeiro passo da análise consistiu no recebimento, organização, atribuição dos Dataframes e verificação da conformidade das informações disponíveis. Após isso, iniciou-se o processo de filtragem dos dados unicamente para o curso de Ciência da Computação e, em seguida, das colunas que seriam utilizadas para realizar o estudo, as quais foram selecionadas:


Durante a seleção, foi identificado que o DataFrame correspondente ao SISU 2018.1 possuía um erro de digitação na coluna 'nome_curso', na qual estava escrito 'CIĘNCIA DA COMPUTAÇĂO'. Com o intuito de resolver esse empecilho, houve a renomeação dos dados dessa coluna para 'CIÊNCIA DA COMPUTAÇÃO', fazendo com que a filtragem ocorresse de forma adequada. Frente a isso, foram construídas visualizações para melhor compreensão dos dados extraídos, tal quais gráficos de barra e boxplots para se entender a nivelação e distribuição das notas dos candidatos aprovados.

````python
# Importando as bibliotecas utilizadas para a construção do código
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Importando os csv's do SISU correspondente aos anos 2018 a 2022, as duas edições de cada ano.
# No entanto, os csv's disponiblizados dos anos de 2021 e 2022 são apenas da primeira edição.
urls_2018_1 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2018/sisu_ufcg_2018_1.csv'
urls_2018_2 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2018/sisu_ufcg_2018_2.csv'

urls_2019_1 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2019/sisu_ufcg_2019_1.csv'
urls_2019_2 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2019/sisu_ufcg_2019_2.csv'

urls_2020_1 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2020/sisu_ufcg_2020_1.csv'
urls_2020_2 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2020/sisu_ufcg_2020_2.csv'

urls_2021_1 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2021/sisu_ufcg_2021.csv'

urls_2022_1 = 'https://raw.githubusercontent.com/Guardians-DSC/dados/main/SISU/sisu_2022/sisu_ufcg_2022.csv'
````

````python
# Atribuindo os df's
Sisu2018_1 = pd.read_csv(urls_2018_1)
Sisu2018_2 = pd.read_csv(urls_2018_2)
Sisu2019_1 = pd.read_csv(urls_2019_1)
Sisu2019_2 = pd.read_csv(urls_2019_2)
Sisu2020_1 = pd.read_csv(urls_2020_1)
Sisu2020_2 = pd.read_csv(urls_2020_2)
Sisu2021_1 = pd.read_csv(urls_2021_1)
Sisu2022_1 = pd.read_csv(urls_2022_1)
````

````python
display(Sisu2022_1) # Visualização de um dos dataframe para verificar se o carregamento foi efetuado
````

````python
Sisu2022_1.columns.tolist() # Visualizando as colunas presentes no df, com o objetivo de identificar quais colunas seriam analisadas
# As escolhidas foram: 'ano' , 'edicao' , 'nome_curso' , 'tipo_mod_concorrencia' , 'nota_l', 'nota_ch', 'nota_cn' , 'nota_m' , 'nota_r' , 'nota_candidato' , 'nota_corte' , 'aprovado'
````

````python
# Criação de função para filtrar de acordo com as colunas escolhidas e aprovados
def filtrando(df):
  curso = ['CIÊNCIA DA COMPUTAÇÃO']

  frame = pd.DataFrame(df,columns=['ano', 'edicao' , 'nome_curso' , 'tipo_mod_concorrencia' , 'nota_l', 'nota_ch', 'nota_cn' , 'nota_m' , 'nota_r' , 'nota_candidato' , 'nota_corte', 'aprovado'])
  # Seleciona apenas as colunas que serão analisadas
  df_filtrado = frame[df['nome_curso'].isin(curso)]
  df_filtradof = df_filtrado.loc[df_filtrado['aprovado'] == 'S']
  df_filtradofs = df_filtradof.loc[df_filtrado['tipo_mod_concorrencia'] == 'A']
  #Analisando apenas o comportamento das notas na ampla concorrência
  return df_filtradofs

# Alterando dado do csv de 2018_1, a fim de padronizar a nomeação do curso para facilitar a filtragem
Sisu2018_1['nome_curso'] = Sisu2018_1['nome_curso'].replace('CIĘNCIA DA COMPUTAÇĂO', 'CIÊNCIA DA COMPUTAÇÃO')
````

````python
# Filtrando apenas para o curso de CC por meio da função criada acima
cc2018_1 = filtrando(Sisu2018_1)
cc2018_2 = filtrando(Sisu2018_2)
cc2019_1 = filtrando(Sisu2019_1)
cc2019_2 = filtrando(Sisu2019_2)
cc2020_1 = filtrando(Sisu2020_1)
cc2020_2 = filtrando(Sisu2020_2)
cc2021_1 = filtrando(Sisu2021_1)
cc2022_1 = filtrando(Sisu2022_1)
````

````python
display(cc2018_1) # Checando se a filtragem funcionou de forma correta
pd.set_option('display.max_row', 10) # ajustando a visualização de linhas
````

A partir da visualização do gráfico 1, é notório que a nota média de linguagens tende a variar entre o intervalo de 600.0 a 650.0 pontos, sendo maior na primeira edição do SISU. Em complemento, é perceptível que as maiores notas da área ao longo dos anos não ultrapassa os 760 pontos, fato que pode ser entendido através de duas lentes: a baixa valoração que o sistema de avaliação do ENEM (Teoria de Resposta ao Item) designa a essa área e ao baixo enfoque direcionado dos alunos para/com esse escopo de estudo, que pode ser motivado por razões intrínsecas ou pelo entendimento da TRI, se estabelecendo como uma relação de casualidade reversa.
Dentre os anos analisados, a edição do SISU que apresentou a maior média foi a de 2019.1, sendo ela 651.11 e a menor foi na segunda edição de 2018.2, a qual atingiu 600.51 como média. Também cabe destacar que o ano de 2019.2 possui a maior nota máxima entre as edições analisadas, o que se mostra dissintônico da tendência de redução das notas que é identificada na segunda edição em comparação a primeira, entretando, ao observar os pontos da nota média e mínima, supõe-se que o dado observado trata-se de um valor atípico, já que esses outros seguem a redução esperada. Em acréscimo, é visto que a edição de 2020.1 possui a maior nota mínima do conjunto analisado, mas que não se mostra muito distante dos demais valores das edições seguintes e anteriores.

# Conclusões

A distribuição do desempenho das notas do Enem dos candidatos aprovados em Ciência da Computação na UFCG pelo SISU mostra-se inclinada para as áreas de Redação e Matemática. Tal inclinação pode ser justificada em vista da alta possibilidade de crescimento de classifiçação quantitativa nas vagas disponibilizadas pelo SISU - tendo em mente que matemática tem a nota mais valorizada pela Teoria de Resposta ao Item (TRI) e também pela objetividade da produção e avaliação da Redação - e pela correlacionalidade dessas áreas com o escopo de estudo da graduação. Ademais, vale-se destacar o crescimento quase contínuo da nota de corte ao decorrer das edições SISU/Enem, o que indica a popularização do curso entre os estudantes, mas também faz-se menção ao período atípico causado pela pandemia de COVID-19, que refletiu na redução temporária da nota de corte de Computação.

Diante dos resultados obtidos, vale-se compreender também a relação mais específica do desempenho das áreas em destaque com os interesses intrísecos dos canditatos, de maneira a fortalecer e incentivar o estudo desta e das outras áreas - Linguagens, Ciências Humanas e Ciências da Natureza - para alcançar uma maior média geral e também aumentar o arcabouço de conhecimento de mundo do estudante. Além disso, é fundamental entender como fatores identitários podem influenciar na escolha do curso e no desempenho entre as áreas, como recortes de gênero e raça. Destarte, a partir da identificação desses grupos particulares, poder-se-á incentivar profundamente um estudo de qualidade para as áreas do Enem, facilitando assim, a entrada e permanência desses alunos ao curso de Ciência da Computação na UFCG.
